```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family']='serif'
rcParams['font.size']=16
```

```python
L = 1
nt = 100
nx = 51
alpha = 1.22e-3
dx = L/(nx-1)

Ti = numpy.zeros(nx)
Ti[0] = 100
```

```python
def ftcs(T,nt,dt,dx,alpha):
    for n in range(nt):
        Tn= T.copy()
        T[1:-1] = Tn[1:-1] + alpha*dt/dx**2*(Tn[2:] - 2*Tn[1:-1] + Tn[:-2])
    return T
```

```python
sigma = 1/2.0
dt = sigma*dx*dx/alpha
T = ftcs(Ti.copy(),nt,dt,dx,alpha)
```

```python
pyplot.plot(numpy.linspace(0,1,nx),T, color='#901060', lw=2, ls='-')
pyplot.ylim(0,100)
pyplot.xlabel('len of rod')
pyplot.ylabel('Temp');
```

```python
nt = 1000
T = ftcs(Ti.copy(),nt,dt,dx,alpha)
```

```python
pyplot.plot(numpy.linspace(0,1,nx),T, color = '#054545',ls='-',lw=3)
pyplot.ylim(0,100)
pyplot.xlabel('len rod')
pyplot.ylabel('temp');
```

```python
def ftcs_mixed(T,nt,dt,dx,alpha):
    for n in range(nt):
        Tn = T.copy()
        T[1:-1]=Tn[1:-1] + alpha*dt/dx**2*(Tn[2:] - 2*Tn[1:-1] + Tn[:-2])
        T[-1] = T[-2]
    return T
```

```python
nt = 1500
T = ftcs_mixed(Ti.copy(),nt,dt,dx,alpha)
pyplot.plot(numpy.linspace(0,1,nx),T, color='#758389',ls='-',lw=3)
pyplot.ylim(0,100)
pyplot.title('rod w/  L dirichlet dc, R neumann bc\n')
```
